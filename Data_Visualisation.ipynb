```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
df=pd.read_csv('tips.csv')
df.head()
```

```python
## Scatter plot
plt.scatter(x='tip',y='total_bill',data=df,c='g',label='bill') #c Is for the color of the plot
plt.scatter(x='tip',y='size',data=df,c='r',marker='+',label='size') ## Marker is for how the marker will look like
plt.legend() ## This willmake the  dot and the plus sugn to be visible in the graph
plt.legend(title='Scatter Plot')
plt.xlabel('tip',c='g',fontsize=20)
plt.ylabel('bill and size',c='g',fontsize=20)
```
